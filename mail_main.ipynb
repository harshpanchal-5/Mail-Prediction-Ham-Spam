```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from  sklearn.metrics import accuracy_score
```

**Data Collection & Pre-processing**

```python
df= pd.read_csv('mail_data.csv')
df.tail()
```

```python
#replace the null value into null striong

df=df.where((pd.notnull(df)),'')
```

```python
df.shape
```

```python
# Changing the value Spam-->0, Ham-->1
df['Category'] = df['Category'].map({'spam': 0, 'ham': 1})

# Show converted values
print("Converted Categories:", df['Category'].unique()) 
```

```python
df.head()
```

```python
# Seprating the Text & Label
X=df["Message"]
Y=df['Category']
```

```python
print(X)
print(Y)
```

```python
#Splittng Data Into Test And Train
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#Feature Extraction   Transform Data   Text--->Numerical

feature_extraction=TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
```

```python
X_train_feat = feature_extraction.fit_transform(X_train)
X_test_feat = feature_extraction.transform(X_test)


# Converting Y_train, Y_test into integer for easy understanging for model.
Y_test= Y_test.astype('int')
Y_train= Y_train.astype('int')
```

```python
print(X_train_feat)
```

```python
model= LogisticRegression()
```

```python
#train the logistic regression with the help of data..
model.fit(X_train_feat,Y_train)
```

```python
#Predection on the training data..

predict_train= model.predict(X_train_feat)
Acc_train_data= accuracy_score(Y_train, predict_train)
print(Acc_train_data) 
```

```python
predict_test = model.predict(X_test_feat)
Acc_test_data = accuracy_score(Y_test, predict_test)
print("Test Accuracy:", Acc_test_data)
```

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
